```python
answer = 3
guess = 0

while guess != answer:
    guess = int(input('please make guess during 1~6.: '))
    if guess > answer:
        print('bigger than answer')
    elif guess < answer:
        print('smaller than answer')
    else:
        print('bingo')
```

```python
answer = 3
guess = 0

while True:
    guess = int(input('please make guess during 1~6.: '))
    if guess > answer:
        print('bigger than answer')
    elif guess < answer:
        print('smaller than answer')
        break
    else:
        print('bingo')
        break
```

```python
i = 0
num_sum = 0
while i <= 100:
    if i % 2 == 1:
        i = i + 1
        continue
    num_sum = num_sum + i
    i = i + 1
print('2+4+6......+100 = {}'.format(num_sum))
```
