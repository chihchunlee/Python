```python
# 用help查詢
import sys
help(sys)
```

```python
# 查詢可用方法
dir(sys)
```

```python
# 查詢方法的功能用法
help(sys.exit)
```

```python
# ?sys.exit 跳出額外視窗顯示
?sys.exit
```

```python
# 查詢哪些功能
l = [1,2,3,4,5]
dir(l)
```

```python
# 查功能怎麼使用 append
help(l.append)
```

```python
# 實測
l.append(2)
print(l)
```

```python
# 查功能怎麼使用 count
help(l.count)
```

```python
# 實測
l.count(2)
```

```python
# 跳出額外視窗顯示
?l.count
```
